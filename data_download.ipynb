```python
'''
It splits a filename into 2 parts:
[0] the name before the dot
[1] the extension (including the dot .
'''
import os

def rename_files(folder):
    for i, file in enumerate(os.listdir(folder), start=0): #enumrate position + value
        ext = os.path.splitext(file)[1]  #split mane, extentn
        os.rename(
            os.path.join(folder, file), 
            os.path.join(folder, f"preview_{i}{ext}")
        )

# Example:
rename_files(r"C:\Users\aliza\Desktop\og_behaviour\dataset_2\howling")
```

```python
#arrange 
import os

def rename_sequential(folder, prefix="preview_", ext=".mp4"):
    files = [f for f in os.listdir(folder) if f.endswith(ext)]
    files.sort()  # Ensure correct order

    for i, filename in enumerate(files, start=0):
        old_path = os.path.join(folder, filename)
        new_name = f"{prefix}{i}{ext}"
        new_path = os.path.join(folder, new_name)
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} → {new_name}")

# Example usage:
rename_sequential(r"C:\Users\aliza\Desktop\og_behaviour\dataset_2\panting")
```

```python
#working without need of proxy  shorts
import yt_dlp
import re
import os
import requests

search_query = "dog sitting"   
max_results = 100            
download_folder = "sitting_downloads_yt"
#cookies_file = "cookies (1).txt"  


os.makedirs(download_folder, exist_ok=True)


print(f"Searching for Shorts: {search_query}")
search_url = f"https://www.youtube.com/results?search_query={search_query.replace(' ', '+')}+shorts"
html = requests.get(search_url).text


video_ids = list(set(re.findall(r"watch\?v=(\S{11})", html)))
print(f"Found {len(video_ids)} videos total.")

video_ids = video_ids[:max_results]

ydl_opts = {
    "format": "best[ext=mp4][vcodec^=avc1][acodec^=mp4a]/best",
    "outtmpl": f"{download_folder}/%(title)s.%(ext)s",
    #"cookiefile": cookies_file
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    for vid_id in video_ids:
        url = f"https://www.youtube.com/watch?v={vid_id}"
        print(f"⬇ Downloading: {url}")
        ydl.download([url])

print("All downloads complete!")
```
